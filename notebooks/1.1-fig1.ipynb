```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import norm, gamma

# 1. Setup the figure and style
sns.set_style("white") # Clean background
fig, axes = plt.subplots(1, 2, figsize=(4.2, 2), sharey=False)

# 2. Generate Data for the Plots
x = np.linspace(0, 10, 500)

# -- Left Plot Data (Narrow Gaussian peaked at the end) --
# Mean=8 (Ripe), Scale=0.5 (Narrow)
y_left = norm.pdf(x, loc=8, scale=0.5) 

# -- Right Plot Data (Skewed distribution) --
# Using a Gamma distribution to get that left-leaning skew
y_right = gamma.pdf(x, a=2.5, loc=0, scale=2)

# 3. Create the Left Plot (Probability vs Ripeness)
color_left_border = "#1f4e5f" # Dark Teal
color_left_fill = "#dbe9f0"   # Light Blue-Grey

sns.lineplot(x=x, y=y_left, ax=axes[0], color=color_left_border, linewidth=3.5)
axes[0].fill_between(x, y_left, color=color_left_fill, alpha=1)

# Styling Left Axis
axes[0].set_ylabel("probability", fontsize=14, color='black')
axes[0].set_xlabel("")
# Custom ticks to match "Unripe" -> "Ripe"
axes[0].set_xticks([1, 8]) 
axes[0].set_xticklabels(["unripe", "ripe"], fontsize=14)
axes[0].set_yticks([]) # Hide y-axis numbers
axes[0].set_ylim(0, max(y_left)*1.1)

# 4. Create the Right Plot (Reward Distribution)
color_right_border = "#8c5128" # Brown
color_right_fill = "#f0dcca"   # Peach/Beige

sns.lineplot(x=x, y=y_right, ax=axes[1], color=color_right_border, linewidth=3.5)
axes[1].fill_between(x, y_right, color=color_right_fill, alpha=1)

# Styling Right Axis
axes[1].set_ylabel("")
axes[1].set_xlabel("")
# Custom ticks to match "Low reward" -> "High"
axes[1].set_xticks([1, 9])
axes[1].set_xticklabels(["low reward", "high reward"], fontsize=14)
axes[1].set_yticks([])
axes[1].set_ylim(0, max(y_right)*1.1)

# 5. Global Styling and the "X" symbol
# Remove top and right spines for a clean look
sns.despine(left=False, bottom=False)

# Thicken the visible axes lines to match the sketch style
for ax in axes:
    ax.spines['left'].set_linewidth(3)
    ax.spines['bottom'].set_linewidth(3)
    ax.spines['left'].set_color('black')
    ax.spines['bottom'].set_color('black')

# Add the multiplication symbol in the middle
# We place this relative to the figure (0.5 is the horizontal center)
# fig.text(0.5, 0.5, r'$\times$', fontsize=40, ha='center', va='center', weight='bold')

plt.tight_layout()
plt.subplots_adjust(wspace=0.3) # Adjust space between plots to fit the X
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# 1. Setup the data
# We generate two normal distributions.
# 'loc' changes the position (mean), 'scale' changes the width (std dev).
np.random.seed(42) # For reproducibility
data_left = np.random.normal(loc=0, scale=5.5, size=10000)
data_right = np.random.normal(loc=15, scale=4.5, size=10000)

# 2. Create the figure
plt.figure(figsize=(12, 5))

# 4. Create the Right Plot (Reward Distribution)
color_right_border = "#8c5128" # Brown
color_right_fill = "#f0dcca"   # Peach/Beige

# 3. Plot the curves using Seaborn's kdeplot
# Left Curve (Lighter Pink)
sns.kdeplot(
    data_left, 
    color="#fcc39b",  # Light pink hex code
    fill=True,        # Fills the area under the curve
    alpha=0.3,        # Transparency (0 to 1)
    linewidth=10,       # Thickness of the outline
    bw_adjust=3.5,
        # Smoothing parameter
)

# Right Curve (Darker Magenta)
sns.kdeplot(
    data_right, 
    color="#f57a23",  # Deep magenta hex code
    fill=True, 
    alpha=0.5, 
    linewidth=10,
    bw_adjust=3.5
)

# 4. Styling to match the image
# Remove the box frame and axis labels for that clean look
sns.despine(left=True, bottom=True) # Keep bottom line (spine) like the image
plt.yticks([]) # Hide Y-axis numbers
plt.xticks([]) # Hide X-axis numbers
plt.xlabel('') # Remove X-axis label
plt.ylabel('') # Remove Y-axis label
plt.xlim(-18, 35)

# Optional: Add a flat baseline if the bottom spine isn't enough
# plt.axhline(0, color='gray', linewidth=2) 
# make background transparent
plt.gca().patch.set_alpha(0.0)

# doesn't work in notebook
plt.savefig('kde_plot.png', transparent=True, dpi=300)
```
